```python
from google_play_scraper import app, reviews, Sort, reviews_all
import pandas as pd
pd.options.mode.chained_assignment = None
import datetime as dt
import csv 
```

```python
scrapingreview = reviews_all(
    'id.dana',         
    lang='id',             
    country='id',          
    sort=Sort.MOST_RELEVANT           
)
```

```python
df_reviews = pd.DataFrame(scrapingreview)
jumlah_ulasan, jumlah_kolom = df_reviews.shape
print(f'Jumlah ulasan: {jumlah_ulasan}')
print(f'Jumlah Atribut: {jumlah_kolom}')
```

```python
clean_review_df = df_reviews.dropna()
clean_review_df = clean_review_df.drop_duplicates()
jumlah_ulasan_setelah_hapus_duplikat, jumlah_kolom_setelah_hapus_duplikat = clean_review_df.shape
print(f'Jumlah ulasan setelah data dibersihkan: {jumlah_ulasan_setelah_hapus_duplikat}')
print(f'Jumlah Atribut setelah data dibersihkan: {jumlah_kolom_setelah_hapus_duplikat}')
```

```python
clean_review_df.to_csv('dataset_aplikasi_dana.csv', index=False)
print("Data berhasil disimpan ke dalam file 'dataset_aplikasi_dana.csv'")
```
